```python
#1. A system checks if a user is eligible to vote based on their age.
# Write logic to ask the user for their age and determine if they are eligible to vote based on whether they are 18 or older.

A=float(input("Enter your age:"))
if A>=18:
    print("You are eligible to vote")
else:
    B=18-A
    print("You need to wait for {} years to cast your vote".format(B))
```

```python
#2. A program processes a list of numbers and needs to find the largest value. Write logic to identify and return the largest number from a given list.
```

```python
# Method 1
numbers=[10,25,15,89,64,30]
Largest=numbers[0]
for num in numbers:
    if num>Largest:
        Largest=num
print("The largest number from the given list is {}".format(Largest))
```

```python
# Method 2
numbers=[10,25,15,89,64,30]
Largest=max(numbers)
print("The largest number from the given list is {}".format(Largest))
```

```python
# 3. A company provides employees with a 10% bonus if their salary exceeds $50,000. Write logic to determine the bonus amount based on the given salary.
Sal=int(input("Enter your salary:"))
if Sal>50000:
    Bonus=int(Sal*0.10)
    print(f"You are eligible and your bonus amount is:{Bonus}")
else:
    print("You are not eligible for bonus")
```

```python
# 4. A program evaluates a number to determine if it is even or odd. Write logic to check whether a given number is even or odd
A=int(input("Enter the number: "))
if A%2==0:
    print(f"{A} is an Even number.")
else:
    print(f"{A} is an Odd number.")
```

```python
# 5. A text-processing tool reverses a given word or sentence for formatting purposes. Write logic to take a word or sentence as input and produce its reversed version.
text=input("Enter any word or sentence: ")
print("The original text: ", text)
print("The reversed text: ",text[::-1])
```

```python
# 8. A banking system processes withdrawal requests and ensures the user has enough balance. Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.
Acc_Bal=20000
Withdraw=int(input("Enter withdraw amount: "))
if Withdraw < Acc_Bal:
    Balance=Acc_Bal-Withdraw
    print(f"Withdrawl successfull. Your account balance is: {Balance}")
else:
    print("Insufficient funds tp process Withdrawl")
```

```python
# 9. A calendar system verifies whether a given year is a leap year based on standard leap year rules.
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a Leap Year")
else:
    print(year, "is Not a Leap Year")
```
